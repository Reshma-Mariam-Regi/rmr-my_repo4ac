```python
import cv2
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Verify OpenCV
print(f"OpenCV version: {cv2.__version__}")

# Verify TensorFlow
print(f"TensorFlow version: {tf.__version__}")

# Verify Numpy
print(f"Numpy version: {np.__version__}")

# Test OpenCV with a simple image read
img = np.zeros((100, 100, 3), dtype=np.uint8)  # Create a black square image
cv2.putText(img, 'Test', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('OpenCV Test Image')
plt.show()
```
