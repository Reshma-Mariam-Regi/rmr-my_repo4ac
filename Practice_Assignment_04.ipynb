```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('credit_card_default.csv')
data.head()
```

```python
data.drop(['ID', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('default payment next month', axis = 1)
y = data['default payment next month']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# kNN 

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train_scaled, y_train)
knn_pred = knn.predict(x_test_scaled)
```

```python
accuracy_score(y_test,knn_pred)
```

```python
# SVM

from sklearn.svm import SVC
svm = SVC()
```

```python
svm.fit(x_train_scaled, y_train)
svm_pred = svm.predict(x_test_scaled)
```

```python
accuracy_score(y_test,svm_pred)
```

```python
# Gradient Boosting

from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
```

```python
gb.fit(x_train_scaled, y_train)
gb_pred = gb.predict(x_test_scaled)
```

```python
accuracy_score(y_test, gb_pred)
```

```python
# AdaBoost

from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
```

```python
ada.fit(x_train_scaled, y_train)
ada_pred = ada.predict(x_test_scaled)
```

```python
accuracy_score(y_test, ada_pred)
```

```python
# XG Boost
!pip install xgboost

from xgboost import XGBClassifier
xgb = XGBClassifier()
```

```python
xgb.fit(x_train_scaled, y_train)
xgb_pred = xgb.predict(x_test_scaled)
```

```python
accuracy_score(y_test, xgb_pred)
```

```python
# hyperparameter tuning

# tuning on KNN

from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']}
```

```python
grid_search_knn = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search_knn.fit(x_train_scaled, y_train)
```

```python
print("Best Parameters:", grid_search_knn.best_params_)
print("Best Score:", grid_search_knn.best_score_)
```

```python
# tuning on xgboost

param_grid1 = {'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7]}
```

```python
grid_search_xgb = GridSearchCV(xgb, param_grid1, cv=5, scoring='accuracy')
grid_search_xgb.fit(x_train_scaled, y_train)
```

```python
print("Best Parameters:", grid_search_xgb.best_params_)
print("Best Score:", grid_search_xgb.best_score_)
```
