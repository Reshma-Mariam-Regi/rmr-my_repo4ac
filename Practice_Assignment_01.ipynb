# 5. Perform meaningful visualizations using the dataset. Bring at least 3             visualizations.


```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
plt.hist([diabetes[diabetes['Outcome'] == 0]['Age'], diabetes[diabetes['Outcome'] == 1]['Age']],
         color=['orange', 'black'], label=['Non-Diabetic', 'Diabetic'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram of Ages for Diabetic and Non-Diabetic Patients')
plt.show()
```

```python
underweight = diabetes[diabetes['BMI'] < 18.5]
healthy_range = diabetes[(diabetes['BMI'] >= 18.5) & (diabetes['BMI'] < 25)]
overweight = diabetes[diabetes['BMI'] >= 25]

plt.figure(figsize=(10, 6))
sns.kdeplot(underweight['Age'], label='Underweight', fill=True)
sns.kdeplot(healthy_range['Age'], label='Healthy Range', fill=True)
sns.kdeplot(overweight['Age'], label='Overweight', fill=True)

plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Diabetes for BMI Categories')
plt.legend()

plt.show()
```
