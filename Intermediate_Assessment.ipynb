# Pre - processing

```python
tr.previous_year_rating.value_counts()
```

```python
tr['previous_year_rating'].fillna(tr['previous_year_rating'].median(),inplace = True)
```

```python
tr.isna().sum()
```

```python
def impute(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)

    iqr = q3 - q1
    
    up_lim = q3 + 1.5 * iqr
    low_lim = q1 - 1.5 * iqr
    
    print("IQR = ",iqr, " Upper limit = ",up_lim, " Lower limit = ",low_lim)
    
    return column.apply(lambda x: up_lim if x > up_lim else(low_lim if x < low_lim else x))
```

```python
tr['no_of_trainings'] = impute(tr['no_of_trainings'])

tr['age'] = impute(tr['age'])

tr['previous_year_rating'] = impute(tr['previous_year_rating'])

tr['length_of_service'] = impute(tr['length_of_service'])
```

```python
plt.boxplot(tr['no_of_trainings'])
plt.title('No. of trainings')
plt.show()

plt.boxplot(tr['age'])
plt.title('Age')
plt.show()

plt.boxplot(tr['previous_year_rating'])
plt.title('Previous year ratings')
plt.show()

plt.boxplot(tr['length_of_service'])
plt.title('Length of service')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = tr.drop('is_promoted', axis = 1)
y = tr.is_promoted
```

```python
x = scaler.fit_transform(x)
```

# Modeling

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25,random_state = 42)
```

```python
from sklearn.metrics import f1_score,accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 100,random_state = 1)
rfc.fit(x_train, y_train)
```

```python
rf_pred = rfc.predict(x_test)
```

```python
accuracy_score(y_test, rf_pred)
```

```python
print('Random Forest F1 Score: ',f1_score(y_test, rf_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []
n = np.arange(3,15)
```

```python
for k in n:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
    classifier.fit(x_train,y_train)
    knn_pred = classifier.predict(x_test)
    acc = accuracy_score(y_test,knn_pred)
    metric_k.append(acc)
```

```python
metric_k
```

```python
plt.plot(n,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('acc')
plt.grid()
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 14, metric = 'euclidean')
classifier.fit(x_train,y_train)
knn_pred = classifier.predict(x_test)
```

```python
accuracy_score(y_test,knn_pred)
```

```python
print('KNN F1 Score: ',f1_score(y_test, knn_pred))
```
